```python
y = y.as_matrix().astype(int)
```

```python
from sklearn.linear_model import SGDClassifier
from sklearn.multiclass import OneVsRestClassifier
```

```python
Z = OneVsRestClassifier(SGDClassifier(loss="log", n_iter=50))
```

```python
# initial test of data on SGCClassifier
Z.fit(X, y.astype(bool))
```

```python
import numpy as np
# examine the coefficients
pd.DataFrame(zip(X.columns, np.transpose(Z.coef_)))
```

```python
# Splitting data 70/30 train/test 
# splits file
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
# checking splits
X.shape, X_train.shape, X_test.shape
```

```python
# checking splits
y.shape, y_train.shape, y_test.shape
```

```python
# normalising data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
# note generate mean and stdev against training, which provides you with a normalising fit for future test data 
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# let's train the model now with the training data
Z = OneVsRestClassifier(SGDClassifier(loss="log", n_iter=50))
```

```python
Z.fit(X_train, y_train.astype(bool))
```

```python
# Evaluating model 
Y_predicted_test = Z.predict(X_test)
```

```python
Y_predicted_test.mean()
```

```python
from sklearn.metrics import f1_score 
score = f1_score(y_test, Y_predicted_test, average="micro")
```
